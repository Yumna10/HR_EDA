```python
df = df.drop(columns=['Unnamed: 14', 'Unnamed: 15'])
```

# What are the salaries of PayType of each department?

```python
avg = df['Salary'].mean().round(2)

salary_avg = df[df['Salary'] > avg]

pd.pivot_table(
    data = salary_avg,
    index = 'Dep',
    columns = 'PayType',
    values = 'Salary',
    aggfunc = ['sum', 'mean']
)
```

# What is the gender distribution across different PayType and their salaries?

```python
pd.pivot_table(
    data = df,
    index = 'Gender',
    columns = 'PayType',
    values = 'Salary',
    aggfunc = ['sum', 'mean', 'count']
)
```

# What is the gender distribution across different branches?

```python
pd.pivot_table(
    data = df,
    index = 'Branch',
    columns = 'Gender',
    values = 'Salary',
    aggfunc = ['count']
)
```

# What is the gender distribution across different departments?


```python
recent_hired = df[df['Hire date'] < '2010-1-1']

pd.pivot_table(
    data = recent_hired,
    index = 'Gender',
    columns = 'Dep',
    values = 'Salary',
    aggfunc = ['count']
)
```

# What is are the salaries of each department based on the position?

```python
highest = df['Salary']

pd.pivot_table(
    data = df,
    index = 'Dep',
    columns = 'Position',
    values = 'Salary',
    aggfunc = ['max','sum','mean','count']
)
```
